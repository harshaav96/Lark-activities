```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\User\Desktop\Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv")
```

```python
data
```

# 1.To compare churn count with respect to gender

```python
sns.countplot(x=data['gender'],hue=data['Churn'],data=data)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender vs Churn count",fontsize=16)
plt.show()
```

# 2.To find the number of female senior citizens in the dataset

```python
data1=data[(data['gender']=='Female') & (data['SeniorCitizen']==1)]
data1['gender'].count()
```

```python
sns.countplot(x=data1['SeniorCitizen'],data=data1)
plt.title('Number of Female Senior Citizens')
plt.grid()
plt.show()
```

# 3.To compare Tenure with Total charges

```python
plt.figure(figsize=(8,6))
plt.scatter(data['tenure'],data['TotalCharges'],s=10)
plt.title('Tenure vs Total charges',fontsize=18)
plt.xlabel('Tenure', fontsize=12)
plt.ylabel('Total charges', fontsize=12)
```

# 4.To find out which contract is preferred by senior citizens

```python
senior_data=data[data['SeniorCitizen']==1]
senior_data
```

```python
plt.figure(figsize=(6,6))
sns.countplot(x=senior_data['Contract'])
plt.title('Contract preferred by Senior citizens',fontsize=18)
```

# 5.Findings on Payment method

```python
plt.figure(figsize=(10,6))
sns.countplot(x=data['PaymentMethod'])
plt.title('Preferred payment methods', fontsize=18)
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

```python
#Conclusion-Electronic check is the most popular payment method.
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=senior_data['PaymentMethod'])
plt.title('Payment methods among senior citizens', fontsize=18)
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

```python
#Conclusion-Most of the senior citizens prefer Electronic check as the payment method.
```
