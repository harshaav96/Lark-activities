# 2.Data visualizations

```python
plt.figure(figsize=(8,8))
sns.barplot(x='country',y='Population in thousands (2017)',data=data.sort_values('Population in thousands (2017)',ascending=False).head(10),color='purple')
plt.title('Top 10 countries based on Population in thousands (2017)',fontsize=16,fontweight='bold')
plt.xlabel('Country',fontsize=15)
plt.ylabel('Population in thousands (2017)',fontsize=14)
plt.xticks(rotation=90)
plt.grid(linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x='country',y='Agricultural production index ',data=data.sort_values('Agricultural production index ',ascending=True).head(10),color='blue')
plt.title('Least 10 countries based on Agricultural production index',fontsize=16,fontweight='bold')
plt.xlabel('Country',fontsize=15)
plt.ylabel('Agricultural production index',fontsize=14)
plt.xticks(rotation=90)
plt.grid(linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.bar(data['Region'], data['International trade: Imports (million US$)'])
plt.title('Region vs International trade: Imports (million US$)', fontweight='bold', fontsize=16,)
plt.xlabel('Region', fontweight='bold', fontsize=12)
plt.ylabel('International trade: Imports (million US$)', fontweight='bold', fontsize=12)
plt.xticks(rotation=90)
plt.grid(linestyle='--')
plt.show()
```

```python
plt.hist(data['GDP per capita (current US$)'],rwidth=0.9,density=True)
plt.title('GDP per Capita',fontweight='bold',fontsize=18)
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
```

# 3.Countries with respect to region available in the dataset

```python
data.groupby("Region")[["country"]].count()
```

# 4.On average which region has highest

# a.GDP per capita (current US$)

```python
GDP=data.groupby("Region")[["GDP per capita (current US$)"]].mean()
GDP.sort_values(by="GDP per capita (current US$)",ascending=False)
```

```python
GDP.nlargest(1,"GDP per capita (current US$)")
```

```python
#Northern America has the highest GDP per capita
```

# b.International trade:imports (million US$)

```python
Imp=data.groupby("Region")[["International trade: Imports (million US$)"]].mean()
Imp.sort_values(by="International trade: Imports (million US$)",ascending=False)
```

```python
Imp.nlargest(1,"International trade: Imports (million US$)")
```

```python
#Northern America has the highest International trade:Imports
```
